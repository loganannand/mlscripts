```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
dataset = pd.read_csv(r"C:\Users\lannand\OneDrive - IB4T Ltd\Desktop\Py\UML\Machine Learning A-Z (Codes and Datasets)\Part 2 - Regression\Section 6 - Polynomial Regression\Python\Position_Salaries.csv")
```

```python
dataset.info()
```

```python
dataset
```

```python
X = dataset.iloc[:, 1:-1]
y = dataset.iloc[:, -1]
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X, y) # Train the model on the entire dataset
```

```python
# Build the matrix of features for polynomial linear regression
poly_reg = PolynomialFeatures(degree=4) # Pick the degree (n), i.e., y = b0 + b1x1 + b2x1**2 + b3x1**3 + ... + bnx1**n
```

```python
X_poly = poly_reg.fit_transform(X)
```

```python
lin_reg_2 = LinearRegression() # call instance of the linear regression class to build the polynomial regressor by using the X_poly var
lin_reg_2.fit(X_poly, y)
```

```python
# plot linear regression model 
plt.scatter(X, y, color='red') # real salaries
plt.plot(X, lin_reg.predict(X), color='blue') # predictions from linear regression model
plt.title('true or false (linear reg)')
plt.xlabel('position level')
plt.ylabel('salar')
plt.show()
```

```python
# Vis the polynomial regression result
plt.scatter(X, y, color='red') # real salaries
plt.plot(X, lin_reg_2.predict(X_poly), color='blue') # predictions from polynomial lin regression model
plt.title('true or false (poly reg model)')
plt.xlabel('position level')
plt.ylabel('salar')
plt.show()
```

```python
print(lin_reg.predict([[6.5]])) # have to input an array 
print(lin_reg_2.predict(poly_reg.fit_transform([[6.5]])))
```
